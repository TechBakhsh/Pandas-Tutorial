```python
titanic=sns.load_dataset("titanic")
titanic.head(5)
```

```python
titanic.to_csv("titanic.csv")
```

```python
titanic.describe()
```

* drop some data 

```python
titanic.drop(["age", "fare"], axis=1)
```
